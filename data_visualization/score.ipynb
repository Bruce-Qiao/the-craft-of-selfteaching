```python
#coding:utf-8

import pandas as pd
from matplotlib import pyplot as plt

#plt.rcParams['font.sans-serif'] = ['SimHei']
#plt.rcParams['axes.unicode_minus'] = False

score = pd.read_csv("score.csv")
score
plt.figure(figsize=(15, 10))
plt.plot(score.模数, score.数学)
plt.plot(score.模数, score.英语)
plt.plot(score.模数, score.语文)
plt.plot(score.模数, score.物理)
plt.plot(score.模数, score.化学)
plt.plot(score.模数, score.政治)
plt.plot(score.模数, score.历史)
plt.plot(score.模数, score.综合)
plt.plot(score.模数, score.体育)
#math_average = [115, 115, 115, 115, 115, 115, 115, 115]
#plt.plot(score.模数, math_average)
plt.title("各模成绩统计")
plt.xlabel("模数")
plt.ylabel("成绩")
plt.legend(["数学", "英语", "语文", "物理", "化学", "政治", "历史", "综合", "体育"])
plt.show()
```
